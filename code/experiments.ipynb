# Average track durations

```python
# Per-genre (per-style) track duration analysis
data_notmixed = analyze.select_notmixed(data)
data_with_durations = analyze.select_with_durations(data)

duration_stats = {}
duration_stats['genre'] = analyze.track_duration_per_genre(data_notmixed, type="genre", genres=data_stats['all_genres'])
duration_stats['genre_only'] = analyze.track_duration_per_genre(data_notmixed, type="genre_only", genres=data_stats['all_genres'])
duration_stats['style'] = analyze.track_duration_per_genre(data_notmixed, type="style", genres=data_stats['all_styles'])

# Not enough data for analysis on tracks annotated by a single style
# duration_stats['style_only'] = analyze.track_duration_per_genre(data_notmixed, type="style_only", genres=data_stats['all_styles'])

# Save results to file
pickle.dump(duration_stats, open(analyze.results_duration, 'wb'))
```

```python
analyze.plot_track_durations(duration_stats['genre'], type="genre")
analyze.plot_track_durations(duration_stats['genre_only'], type="genre_only")
analyze.plot_track_durations(duration_stats['style'], type="style", shorten_stylenames=False)

# Group styles by parent genres
for g in data_stats['all_genres']:
    styles = [(g, s) for (gs, s) in duration_stats['style'] if gs == g]
    stats_g = {}
    for s in styles:
        stats_g[s] = duration_stats['style'][s]
    
    analyze.plot_track_durations(stats_g, type="style", shorten_stylenames=True,
                                 title="Boxplot of track durations (sec.) per style (%s genre)" % g)
```
